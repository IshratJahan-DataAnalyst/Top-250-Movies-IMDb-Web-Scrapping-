```python
from bs4 import BeautifulSoup
import requests,openpyxl
```

```python
url = requests.get('https://www.imdb.com/chart/top/')
url.raise_for_status()

soup= BeautifulSoup(url.text,'html.parser')
```

```python
soup.title
```

```python
soup.title.parent.name
```

```python
soup.p
```

```python
excel = openpyxl.Workbook()
sheet=excel.active
sheet.title='Top 250 Movies'
print(excel.sheetnames)
sheet.append(['Movie Rank','Movie Name','Year','Rating'])
```

```python
try:
    movies= soup.find('tbody',class_= "lister-list").find_all('tr')
    for movie in movies:
        name = movie.find('td',class_="titleColumn").a.get_text()
        rank = movie.find('td',class_="titleColumn").get_text(strip=True).split('.')[0]
        year = movie.find('td',class_="titleColumn").span.text.strip('()')
        rating = movie.find('td',class_="ratingColumn imdbRating").text.strip()
        print(name,rank,year,rating)
        sheet.append([name,rank,year,rating])
except Exception as e:
        print(e)
        
excel.save('IMDB Top 250 Movies.xlsx')
```
